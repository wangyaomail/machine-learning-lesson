```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import sklearn.datasets
```

```python
boston = sklearn.datasets.load_boston()
print(boston.DESCR)
```

```python
print(boston.feature_names, len(boston.feature_names))
```

```python
boston_pd = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_pd
```

```python
plt.rcParams['font.size'] = 8
plt.figure(figsize=[10, 15], dpi=200)
for name, i in zip(boston.feature_names, range(1, 1000)):
    col_data = boston_pd[name]
    plt.subplot(5, 3, i)
    plt.scatter(col_data, boston.target, c=boston.target, cmap='cool', alpha=0.5, s=1)
    plt.title(name)
plt.tight_layout()
plt.show()
```

```python
boston_2_data = boston_pd[['LSTAT']].values
boston_2_target = boston.target
print(boston_2_target.shape, boston_2_data.shape)
```

```python
# x = np.linspace(0, 1, 201)
# y = np.linspace(0, 1, 201)
# x, y = np.meshgrid(x, y)
# disp_input = np.vstack((x.flatten(), y.flatten())).T
disp_input = np.linspace(0, 40, 1000)
disp_input = disp_input.reshape((len(disp_input), 1))
print(disp_input)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(boston_2_data, boston_2_target)
disp_predict = model.predict(disp_input)
boston_predict = model.predict(boston_2_data)

plt.scatter(disp_input, disp_predict, c='black', alpha=0.8, s=1)
plt.scatter(boston_2_data, boston_2_target, c=boston_2_target, cmap='cool', s=1)
plt.title(model)
plt.show()
```

```python
from sklearn.svm import SVR
plt.rcParams['font.size'] = 8
plt.figure(figsize=[10, 10], dpi=200)
for depth_i, i in zip([1,3,5,7,10,50,100], range(1, 100)):
    model = SVR(C=depth_i)
    model.fit(boston_2_data, boston_2_target)
    disp_predict = model.predict(disp_input)
    title = "max_depth={}".format(depth_i)
    plt.subplot(3, 3, i)
    plt.scatter(boston_2_data, boston_2_target, c=boston_2_target, cmap='cool', s=1)
    plt.scatter(disp_input, disp_predict, c='red', alpha=0.8, s=0.3)
    plt.title(model)
plt.tight_layout()
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor
plt.rcParams['font.size'] = 8
plt.figure(figsize=[10, 10], dpi=200)
for depth_i, i in zip([(10,),(100,10),(10,10,10),(100,50,10),(160,120,80,40,20,10)], range(1, 100)):
    model = MLPRegressor(hidden_layer_sizes=depth_i, max_iter=10000)
    model.fit(boston_2_data, boston_2_target)
    disp_predict = model.predict(disp_input)
    title = "max_depth={}".format(depth_i)
    plt.subplot(3, 3, i)
    plt.scatter(boston_2_data, boston_2_target, c=boston_2_target, cmap='cool', s=1)
    plt.scatter(disp_input, disp_predict, c='red', alpha=0.8, s=0.3)
    plt.title(model)
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
plt.rcParams['font.size'] = 8
plt.figure(figsize=[10, 10], dpi=200)
for depth_i, i in zip([1, 3, 5, 7, 100,1000], range(1, 100)):
    model = DecisionTreeRegressor(max_depth=depth_i)
    model.fit(boston_2_data, boston_2_target)
    disp_predict = model.predict(disp_input)
    title = "max_depth={}".format(depth_i)
    plt.subplot(3, 3, i)
    plt.scatter(boston_2_data, boston_2_target, c=boston_2_target, cmap='cool', s=1)
    plt.scatter(disp_input, disp_predict, c='green', alpha=0.8, s=2)
    plt.title(model)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import AdaBoostRegressor
plt.rcParams['font.size'] = 8
plt.figure(figsize=[10, 10], dpi=200)
for depth_i, i in zip([1, 3, 5, 7, 100,1000], range(1, 100)):
    model = AdaBoostRegressor(n_estimators=depth_i)
    model.fit(boston_2_data, boston_2_target)
    disp_predict = model.predict(disp_input)
    title = "n_estimators={}".format(depth_i)
    plt.subplot(3, 3, i)
    plt.scatter(boston_2_data, boston_2_target, c=boston_2_target, cmap='cool', s=1)
    plt.scatter(disp_input, disp_predict, c='green', alpha=0.8, s=2)
    plt.title(model)
plt.tight_layout()
plt.show()
```
