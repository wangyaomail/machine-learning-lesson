```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.sans-serif']=['simhei']
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
movies = pd.read_csv("../data/ml-1m/movies.dat",
                     sep="::",
                     header=None,
                     encoding="iso8859",
                     engine="python")
movies
```

```python
users = pd.read_csv("../data/ml-1m/users.dat",
                     sep="::",
                     header=None,
                     encoding="iso8859",
                     engine="python")
users
```

```python
ratings = pd.read_csv("../data/ml-1m/ratings.dat",
                     sep="::",
                     header=None,
                     encoding="iso8859",
                     engine="python")
ratings
```

```python
user_movie = np.zeros((users.shape[0],movies.shape[0]+200))
for (index, row) , i in zip(ratings.iterrows(), range(1000000)):
    user_movie[row[0]-1, row[1]-1] = row[2]
```

```python
pd.DataFrame(user_movie[0])
```

```python
user_movie.shape
```

```python
import sklearn.metrics.pairwise as pws
user_user = pws.cosine_similarity(user_movie)
```

```python
pd.DataFrame(user_user[0])
```

```python
movie_movie = pws.cosine_similarity(user_movie.T)
```

```python
pd.DataFrame(movie_movie[0])
```
